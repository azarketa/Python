## DEFINING FFT FUNCTION

```python
## PLOTTING FFT AMPLITUDE CURVESplt.ylabel("Non-dimensional amplitude", fontsize=15)
def plottempdataforangle(file_list, angle):

    # plt.xlabel("temp $(sec)$")
    # plt.ylabel("$C_{L}$")
    # plt.grid()
    # plt.title("Angle = {0}".format(angle))

    max_rfft = []

    for file in file_list:
        
        if angle in globals()[file].groups_added.means_group.means_channels.angle_pos:
            angle_index = str(globals()[file].groups_added.means_group.means_channels.angle_pos.index(angle) + 1)            
        else:
            return "Angle value {0} not found on file {1}; ignoring plot; try another angle".format(angle, globals()[file])
        
        if hasattr(globals()[file], "__groups_original__"):            
            temp = globals()[file].__groups_original__.__getattribute__("measurement" + angle_index).channels.kistler_signal_time.data
#             temp_data = globals()[file].__groups_original__.__getattribute__("measurement" + angle_index).channels.corrected_kistler_fx_signal.data*np.sin(angle*np.pi/180) + \
#                         globals()[file].__groups_original__.__getattribute__("measurement" + angle_index).channels.corrected_kistler_fy_signal.data*np.cos(angle*np.pi/180)
            temp_data = globals()[file].__groups_original__.__getattribute__("measurement" + angle_index).channels.corrected_kistler_fx_signal.data*np.cos(angle*np.pi/180) - \
                        globals()[file].__groups_original__.__getattribute__("measurement" + angle_index).channels.corrected_kistler_fy_signal.data*np.sin(angle*np.pi/180)
        else:
            temp = globals()[file].groups_original.__getattribute__("measurement" + angle_index).channels.Kistler_signal_time.data
#             temp_data = globals()[file].groups_original.__getattribute__("measurement" + angle_index).channels.corrected_kistler_fx_signal.data*np.sin(angle*np.pi/180) + \
#                         globals()[file].groups_original.__getattribute__("measurement" + angle_index).channels.corrected_kistler_fy_signal.data*np.cos(angle*np.pi/180)
            temp_data = globals()[file].groups_original.__getattribute__("measurement" + angle_index).channels.corrected_kistler_fx_signal.data*np.cos(angle*np.pi/180) - \
                        globals()[file].groups_original.__getattribute__("measurement" + angle_index).channels.corrected_kistler_fy_signal.data*np.sin(angle*np.pi/180)
            
        temp_data_average = np.average(temp_data)
        temp_data_no_average = temp_data - temp_data_average
        temp_data_no_average *= scsg.windows.hamming(len(temp_data_no_average))
        temp_rfft = (np.fft.rfft(temp_data_no_average)/(len(np.fft.rfft(temp_data_no_average))))
        real_temp_rfft = np.real(np.sqrt(temp_rfft*np.conjugate(temp_rfft)))
        max_rfft.append(np.max(real_temp_rfft))
        # plt.plot(temp, temp_data - temp_data_average, label=globals()[file].__alias__)                            

    return angle, max_rfft
```
